```python
#@title Imports
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.special as sp
import scipy.integrate
from sympy import *
```

```python
a_0 = 0.53

def funcao_onda(x, y, z, n, l, m):
    r = np.sqrt(x**2 + y**2 + z**2)
    theta = np.arccos(z/r)
    phi = np.arctan2(y, x)
    
    harmonicos = sp.sph_harm(m, l, phi, theta)
    laguerre = sp.genlaguerre(n-l-1,2*l+1)(2*r/n*a_0)
    
    R = np.sqrt((2.0/n*a_0)**3*math.factorial(n-l-1)/(2.0*n*math.factorial(n+l)))
    Y = R*np.exp(-r/n*a_0)*(2*r/n*a_0)**l*harmonicos*laguerre
    Y = np.nan_to_num(Y)
    return Y
```

```python
delta = 0.5

x_ = np.arange(-20, 20, delta)
y_ = np.arange(-20, 20, delta)
z_ = np.arange(-20, 20, delta)

x, y, z = np.meshgrid(x_, y_, z_)
```

```python
n = 2 
l = 1
m = 0

funcao_plot = funcao_onda(x, y, z, n, l, m)
funcao_plot = abs(funcao_plot)**2

r = np.sqrt(x**2 + y**2 + z**2)

fig, ax = plt.subplots()
plt.subplots_adjust(left=0.15, bottom=0.15)
plot = plt.imshow(np.transpose(funcao_plot[int((0-(-20))/delta)]), cmap='inferno')
cbar = fig.colorbar(plot, ticks=[0])
cbar.ax.set_yticklabels(['$-$'])
cbar.ax.tick_params(labelsize=20)
plt.axis('off')
ax.set_title('n='+str(n)+', l='+str(l)+', m='+str(m))
plt.show()
```

```python
def R(r, n, l):
    
    laguerre = sp.genlaguerre(n-l-1,2*l+1)(2.0*r/(n*a_0))
    R = np.sqrt((2.0/(n*a_0))**3*math.factorial(n-l-1)/(2.0*n*math.factorial(n+l)))*np.exp(-2.0*r/(n*a_0)/2)*(2.0*r/(n*a_0))**l*laguerre

    return r**2*abs(R)**2


n = 2
l = 1

L_1 =[]
r_ = np.arange(0, 20, 0.5)

for i in r_:
    L_1.append(R(i, n, l))

r_a = []
for i in r_:
    r_a.append(i/a_0)

plt.plot(r_a, L_1, label='n = '+str(n)+', l = '+str(l)+'', color='orange')


n = 1
l = 0

L_2 =[]
r_ = np.arange(0, 20, 0.5)

for i in r_:
    L_2.append(R(i, n, l))

r_a = []
for i in r_:
    r_a.append(i/a_0)

plt.plot(r_a, L_2, label='n = '+str(n)+', l = '+str(l)+'', color='green',linestyle='dotted')


n = 3
l = 2

L_3 =[]
r_ = np.arange(0, 20, 0.5)

for i in r_:
    L_3.append(R(i, n, l))

r_a = []
for i in r_:
    r_a.append(i/a_0)

plt.plot(r_a, L_3, label='n = '+str(n)+', l = '+str(l)+'', color='blue', linestyle='--')


plt.legend()
plt.xlabel('r/$a_0$')
plt.ylabel('$r^2|R(r)|^2$')
plt.title('Densidade radial')
plt.show()
```
