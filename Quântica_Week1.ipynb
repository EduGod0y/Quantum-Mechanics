```python
%matplotlib inline
import math
import numpy as np
from numpy import loadtxt
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
# Atividade 1 - Função densidade de probabilidade
def psi1(x):
    return (2*np.pi)*-0.50*np.exp(-5*np.abs(x-2))/np.sqrt(0.1591549431)/np.sqrt(12.403)
def psi2(x):
    return (3*np.pi)*np.exp(-1*np.abs(x+3))/np.sqrt(1.974)/np.sqrt(44.998)
def psi3(x):
    return (2)*np.exp(-0.5*np.abs(x))/np.sqrt(2)/np.sqrt(4)
def psi4(x):
    return (np.pi)**3*np.exp(-10*np.abs(x))/np.sqrt(96.139)
def psi5(x):
    return (np.pi)**2*3**(-0.5*np.abs(x-1))/np.sqrt(194.818)/np.sqrt(0.91)

x = np.arange(-5,5,0.1)

def graf(psi0,x):
  plt.plot(x,psi0(x)**2)
  plt.title('Densidade de Probabilidade')
  plt.xlabel('x')
  plt.ylabel('$|\Psi(x)|^2$')
  
  plt.show()

  area,err = integrate.quad(lambda x :psi0(x)**2,-np.inf,np.inf)
  xmean,err = integrate.quad(lambda x :x*psi0(x)**2,-np.inf,np.inf)
  x2mean,err = integrate.quad(lambda x :x**2*psi0(x)**2,-np.inf,np.inf)

  print ('area = {:.3f}'.format(area)) #Alterei o número de casas decimais para fazer a normalização (estava tendo problema com os decimais)
  print ('média = {:.3f}'.format(xmean))
  

graf(psi1,x)
graf(psi2,x)
graf(psi3,x)
graf(psi4,x)
graf(psi5,x)
```

```python
# Atividade 2 - O mesmo para psi complexo
def psi_1(x):
    r =  (2*np.pi)**-1*np.exp(-0.2*np.abs(x**2/4))/np.sqrt(0.063)/np.sqrt(1.008)/np.sqrt(2.236)
    return complex(r*np.cos(np.pi/3*x),r*np.sin(np.pi/3*x))

def psi_2(x):
    r = (2*np.pi)**-1*np.exp(-0.5*np.abs(x**2/4))/np.sqrt(0.090)/np.sqrt(0.998)
    return complex(r*np.cos(np.pi/3*x),r*np.sin(np.pi/3*x))

def psi_3(x):
    r = (7*np.pi)**-1*np.exp(-0.75*np.abs(x**2/4))/np.sqrt(0.006)/np.sqrt(0.997)
    return complex(r*np.cos(np.pi/3*x),r*np.sin(np.pi/3*x))

def psi_4(x):
    r = (np.pi)**-0.2*np.exp(-0.5*np.abs(x**2/4))/np.sqrt(2.243)
    return complex(r*np.cos(np.pi/3*x),r*np.sin(np.pi/3*x))

def psi_5(x):
    r = (8*np.pi)**-0.5*np.exp(-2*np.abs(x**2/4))/np.sqrt(0.071)/np.sqrt(0.993)
    return complex(r*np.cos(np.pi/3*x),r*np.sin(np.pi/3*x))
x_arr = np.arange(-5,5,0.1)

def graf_at2(psi,x):
  def prob_dens(x):
    z = psi(x)
    return z.real**2 + z.imag**2
  pp=[]
  for x in x_arr:
      pp.append(prob_dens(x))
  plt.plot(x_arr,pp,c = 'b')
  plt.title('Densidade de Probabilidade')
  plt.xlabel('x')
  plt.ylabel('$|\Psi(x)|^2$')
  
  plt.ylim(0,0.5)
  plt.show()

  area,err = integrate.quad(lambda x :prob_dens(x),-np.inf,np.inf)
  xmean,err = integrate.quad(lambda x :x*prob_dens(x),-np.inf,np.inf)
  x2mean,err = integrate.quad(lambda x :x**2*prob_dens(x),-np.inf,np.inf)
  delta_x = np.sqrt(x2mean-xmean**2)
  print ('area = {:.3f}'.format(area)) #Aqui também foi necessário alterar as casas decimais para que a área ficasse normalizada
  print ('média = {:.3f}'.format(xmean))
  print ('desvio padrão = {:.3f}'.format(delta_x))

```

```python
graf_at2(psi_1,x)
graf_at2(psi_2,x)
graf_at2(psi_3,x)
graf_at2(psi_4,x)
graf_at2(psi_5,x)
```

```python
# Calculando a transformada de Fourier
def A(psi,k):
    def integrando1(x):
        z = psi(x)
        return z.real*np.cos(k*x)-z.imag*np.sin(-k*x)
    def integrando2(x):
        z = psi(x)
        return z.real*np.sin(-k*x)+z.imag*np.cos(k*x)    
    A1,err1 = integrate.quad(integrando1,-np.inf,np.inf)
    A2,err2 = integrate.quad(integrando2,-np.inf,np.inf) 
    return complex(A1/(np.sqrt(2*np.pi)),A2/(np.sqrt(2*np.pi)))


dk = 0.1
k_arr = np.arange(-5,5,0.1)

def graf_at2_2(psi,x):
  def prob_dens_fourier(k):
    z = A(psi,k)
    return (z.real**2 + z.imag**2)

  pp_fourier=[]
  for k in k_arr:
      pp_fourier.append(prob_dens_fourier(k))
  plt.plot(k_arr,pp_fourier)
  plt.title('Densidade de Probabilidade k')
  plt.xlabel('k')
  plt.ylabel('$|A(k)|^2$')
  
  plt.ylim(0,1.65)
  plt.show()
  area = np.sum(pp_fourier)*dk
  kmean = np.sum(k_arr*pp_fourier)*dk
  k2mean = np.sum(k_arr**2*pp_fourier)*dk
  delta_k = np.sqrt(k2mean-kmean**2)
  print ('area = {:.3f}'.format(area))
  print ('média = {:.3f}'.format(kmean))
  print ('desvio padrão = {:.3f}'.format(math.sqrt(k2mean-kmean**2)))

graf_at2_2(psi_1,x)
graf_at2_2(psi_2,x)
graf_at2_2(psi_3,x)
graf_at2_2(psi_4,x)
graf_at2_2(psi_5,x)
```
